```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(123456)

from scipy.interpolate import interp1d
upper_out_of_bound = lower_out_of_bound = True
while upper_out_of_bound or lower_out_of_bound:
    steps = 18
    x = np.log(get_population_time(time_rate=0.1, num_time_windows=steps, tmax=10_000_000).tolist())
    y = np.log(sample_population_size(10_000, 10_000_000, steps))
    xnew = np.linspace(x[0], x[-1], num=10000, endpoint=True)
    f_cubic = interp1d(x, y, kind='cubic')
    ynew = f_cubic(xnew)
    upper_out_of_bound = np.sum(np.exp(ynew) > 10_000_000) > 0
    lower_out_of_bound = np.sum(np.exp(ynew) < 10_000) > 0
    
sns.set(style="ticks", font_scale=1.2)
fig, axes = plt.subplots(1, 1, figsize=(8,4), sharex=True, sharey=True)
x_sample = xnew[np.linspace(10, 9999, 60).astype(int)]
y_sample = ynew[np.linspace(10, 9999, 60).astype(int)]

sns.lineplot(x=np.exp(x_sample), y=np.exp(y_sample), ax=axes)
sns.scatterplot(x=np.exp(x_sample), y=np.exp(y_sample), ax=axes)
plt.axvline(x=10, color='k', linestyle='--')
    
plt.yscale('log')
plt.xscale('log')
plt.ylim(1000, 10_000_000)
plt.xlim(1, 10_000_000)

population_time = np.exp(x_sample)
```

```python
demography = np.exp(y_sample)
```

```python
demography = [1e6] * 60
```

```python
np.random.seed(123456)


from sklearn.linear_model import LinearRegression

num_scenario = 1
increment = 0
num_replicates = 100
upper_tree_limit = 500

X = np.array([1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]).reshape(-1, 1)
y = np.array([903272667.6483952, 449680355.2510645,178743721.59285852, 70857015.72050871,26609043.76117958,
              10475248.838195799,4518011.131732858, 2024414.914914666, 1265514.8452846785])
y = np.log(y)
reg = LinearRegression().fit(X, y)

#y = y * 1.5
```

```python

masks = []


for alpha in [1.9, 1.7, 1.5, 1.3]:

    sequence_length = int(np.exp(reg.predict(np.array([[alpha]])).item()))
    print(f"alpha {alpha} sequence length {sequence_length}")

    parameters = sample_parameters(num_scenario,
                                   increment=increment,
                                   num_replicates=num_replicates,
                                   num_time_windows=60,
                                   n_min = 10_000, 
                                   n_max = 10_000_000,
                                   recombination_rates=[1e-8, 1e-8],
                                   population_size = demography,
                                   model = "beta",
                                   alpha = alpha
                                  )

    tree_sequences = simulate_tree_sequence(parameters, population_time, segment_length=sequence_length, seed=None)
    for i in range(0, len(tree_sequences), num_replicates):
        col_events , mask = compute_mask_from_tree_sequences(tree_sequences[i:i+num_replicates], population_time, num_cpus=7, min_coal_tree=30)
    
    ce = col_events / col_events.max()
    ce = pd.DataFrame(ce, columns=np.round(population_time)) 
    mask = pd.DataFrame(mask, columns=np.round(population_time)) 
    
    masks.append([ce, mask])
```
